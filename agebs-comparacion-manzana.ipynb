```python
import pandas as pd
```

```python
# Leer el archivo XLSX
df1 = pd.read_excel('loc-mzna_muestra_ageb_s42_rural_parte01_edos-prioritarios.xlsx', skiprows=1, dtype=str)
```

# Dataframe NO optimizado

# NOTA IMPORTANTE

## df1 : NO optimizado y  df2 : Optimizado

```python
df1
```

```python
# Tomamos todos los cvegeosel_m_l
new_df1 = df1.filter(regex='^cvegeoageb(\.[1-8])?$')
```

```python
new_df1.tail(5)
```

```python
# Apilar las columnas verticalmente 
serie_apilada = new_df1.stack()

# Crear un nuevo DataFrame 
nuevo_df1 = pd.DataFrame(serie_apilada, columns=['cvegeoageb']).reset_index(drop=True)

# Mostrar el nuevo DataFrame
nuevo_df1
```

# Dataframe optimizado

```python
df2 = pd.read_csv('07_marco_muestral_manzana_loc.csv', dtype=str)
```

```python
df2.head(5)
```

```python
# Tomamos todos los cvegeoageb en una sola columna
new_df2 = df2['cvegeoageb']
new_df2 = pd.DataFrame(new_df2)
new_df2
```

# Revisar

```python
#concatenamos
df_concatenado = pd.concat([nuevo_df1, new_df2])
df_concatenado
```

```python
# Eliminamos los valores NaN
df_concatenado_sin_nan = df_concatenado.dropna()

# Identificamos las filas duplicadas
filas_duplicadas = df_concatenado_sin_nan[df_concatenado_sin_nan.duplicated()]

# Eliminamos las duplicadas únicas
filas_duplicadas_unicas = filas_duplicadas.drop_duplicates()

# Verificar si hay filas duplicadas
if not filas_duplicadas_unicas.empty:
    print("Se encontraron filas duplicadas en los dos DataFrames, por lo tanto el dataframe NO optimizado NO está contenido (o parcialmente contenido) en el dataframe optimizado.")
    print(filas_duplicadas_unicas)
else:
    print("No se encontraron filas duplicadas en los dos DataFrames, por lo tanto el dataframe NO optimizado está contenido en el optimizado.")
```

# Join left

```python
resultado_left_join = pd.merge(nuevo_df1, new_df2, on='cvegeoageb', how='left')
resultado_left_join
```
