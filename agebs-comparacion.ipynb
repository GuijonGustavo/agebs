# Revisar

```python
frame1 = new_df1
frame2 = new_df2.to_frame()

# Convierte los DataFrames en MultiIndex utilizando stack
multiindex_df1 = frame1.stack()
multiindex_df2 = frame2.stack()

# Utiliza unstack para convertir el MultiIndex nuevamente en DataFrame
df_unstacked1 = multiindex_df1.unstack()
df_unstacked2 = multiindex_df2.unstack()

# Encuentra los valores duplicados en los DataFrames unstacked
valores_duplicados = df_unstacked1[df_unstacked1.isin(df_unstacked2)].dropna()

# Si hay valores duplicados, imprímelos
if not valores_duplicados.empty:
    print("Se encontraron celdas duplicadas en los dos DataFrames:")
    print(valores_duplicados)
else:
    print("No se encontraron celdas duplicadas en los dos DataFrames.")
```

```python
new_df2 = new_df2.to_frame()
df_concatenado = pd.concat([new_df1, new_df2])

# Eliminamos los valores NaN
df_concatenado_sin_nan = df_concatenado.dropna()

# Identificamos las filas duplicadas
filas_duplicadas = df_concatenado_sin_nan[df_concatenado_sin_nan.duplicated()]

# Verificar si hay filas duplicadas
if not filas_duplicadas.empty:
    print("Se encontraron filas duplicadas en los dos DataFrames, por lo tanto el dataframe NO optimizado NO está contenido (o parcialmente contenido) en el dataframe optimizado.")
    print(filas_duplicadas)
else:
    print("No se encontraron filas duplicadas en los dos DataFrames, por lo tanto el dataframe NO optimizado está contenido en el optimizado.")
```
